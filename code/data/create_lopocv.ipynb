```python
import pandas as pd
import numpy as np
import pickle
import torch
import torch.nn as nn
import argparse
from tqdm import tqdm
import sys
from sklearn.metrics import accuracy_score
import random
import itertools
```

```python
with open('c:\\Users\\hayas\\proj-rank-general\\git\\data\\features_dict_seq.pkl', 'rb') as p:
    features_dict = pickle.load(p)
```

```python
participant_id = []
for i in range(30):
    if len(str(i)) == 1:
        id = "0" + str(i)
    else:
        id = str(i)
    participant_id.append(id)

contents_id = []
for i in range(5):
    contents_id.append(str(i))

print(participant_id)
print(contents_id)
```

```python
lopocv_id_dict = {}
for i in participant_id:
    lopocv_id_dict[i] = {"train_in_test": [], "test": []}

for test in participant_id:
    for j in features_dict.keys():
        if test == j[:2]:
            lopocv_id_dict[str(test)]["test"].append(j)
        else:
            lopocv_id_dict[str(test)]["train_in_test"].append(j)
```

```python
def create_id_dict_per_person(id_list):
    id_dict_per_person = {}
    u_id_set = set()
    for id in id_list:
        u_id_set.add(id[:3])
    u_id_list = sorted(list(u_id_set))
    for u in u_id_list:
        id_dict_per_person[u] = []
        for id in id_list:
            if u == id[:3]:
                id_dict_per_person[u].append(id)

    return id_dict_per_person
```

```python
lopocv_dict = {}
for k, v in lopocv_id_dict.items():
    lopocv_dict[k] = {}
    for k2, v2 in v.items():
        lopocv_dict[k][k2] = []
        id_dict_per_person = create_id_dict_per_person(v2)
        for k3, v3 in id_dict_per_person.items():
            p_list = list(itertools.permutations(v3, 2))
            p_list_del_dup = []
            for p in p_list:
                if (p[1], p[0]) not in p_list_del_dup:
                    p_list_del_dup.append(p)
            lopocv_dict[k][k2].extend(p_list_del_dup)
```
